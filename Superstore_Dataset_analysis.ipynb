````python
#extract Ship Date and Order date column from datetimeIndex
date=pd.DatetimeIndex(df['Order Date'])
date=pd.DatetimeIndex(df['Ship Date'])

#extract month and year from datetimeIndex in Order Month column
df['Order_month']=date.month
df['Order_Year']=date.year

df['Ship_month']=date.month
df['Ship_Year']=date.year
````

````python
df.head()
````

````python
#object dtypes to Categorial
df['Ship Mode']=df['Ship Mode'].astype('category')
df['Customer ID']=df['Customer ID'].astype('category')
df['Customer Name']=df['Customer Name'].astype('category')
df['Country']=df['Country'].astype('category')
df['City']=df['City'].astype('category')
df['State']=df['State'].astype('category')
df['Segment']=df['Segment'].astype('category')
df['Region']=df['Region'].astype('category')
df['Product ID']=df['Product ID'].astype('category')
df['Product Name']=df['Product Name'].astype('category')
df['Order ID']=df['Order ID'].astype('category')
df['Category']=df['Category'].astype('category')
df['Sub-Category']=df['Sub-Category'].astype('category')
df['Ship Date']=df['Ship Date'].astype('datetime64[ns]')
````

**Show the distribution of Sales using Histogram**

````python
#identify figure size of the plot
plt.figure(figsize=(10,8))

plt.hist(df['Sales'], color='Gold', edgecolor='brown')

#add labels and title
plt.xlabel('Sales')
plt.ylabel('frequency')
plt.title('Histogram of Sales')

#show the plot
plt.show()
````



```
More sales fall between $0 and $2000. There are fewest sales after $2000.
```



 **What is the total number of quantity by Ship mode?**

````python
orders_shipmode=df.groupby('Ship Mode')['Quantity'].sum().sort_values(ascending=False)
````

````python
#identify figure size
plt.figure(figsize=(8,8))

orders_shipmode.plot(kind='bar',color='c')

#add label and title
plt.xlabel('Ship mode')
plt.ylabel('Quantity')
plt.title('Histogram of Quantity by Ship mode')

#show the plot
plt.show()
````



```
With more than 12000 quantities sent, Standard Class is the highest ship mode.Followed by Second class mode, which shipped
more than 4000 tons. Last but not least, Same Day mode only dispatched 2000 or fewer volumes.
```

**How quantity profit vary with Order month?**

````python
Profit_monthly=df.groupby('Order_month')['Profit'].sum().reset_index()
````

````python
Profit_monthly
````

````python
#identify figure size of the plot
plt.figure(figsize=(20,12))

# Create the Line Plot
sns.set(style="whitegrid")

df['Order_month'] = df['Order_month'].apply(lambda x:calendar.month_name[x])

Profit_monthly.plot(kind='line', marker='o', label='monthly_profit')
sns.color_palette('Paired')

#add labels and title
plt.xlabel('Order month')
plt.ylabel('Profit')
plt.title('Total monthly profit')

plt.xticks(range(1, 13), calendar.month_abbr[1:])
plt.legend()

#show the plot
plt.show()
````



```
October has the highest profit, surpassing 25,000.September, November, and December all made more
than a 20,000 profit. March has a lower profit total of 5,000.
```
